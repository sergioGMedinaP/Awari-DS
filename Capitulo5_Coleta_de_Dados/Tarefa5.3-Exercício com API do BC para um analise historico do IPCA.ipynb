```python
import requests
from pandas.io.json import json_normalize
import pandas as pd
import matplotlib.pyplot as plt


from datetime import datetime

```

```python
URL = "https://api.bcb.gov.br/dados/serie/bcdata.sgs.16121/dados?formato=json"
df_banco_central = pd.read_json(URL)
df_banco_central.shape
```

```python
df_bc1= df_banco_central.head(50)
```

```python
print(df_bc1)
```

```python
df_x   = df_bc1['data']
df_y = df_bc1['valor']


plt.figure( figsize=(30, 5)) # alterar tamanho
plt.plot(df_x,df_y)
plt.xlabel('Anos')
plt.ylabel('Percentual')
plt.title("Relação de IPCA no periodo 1981 -1995")
plt.show()

```
